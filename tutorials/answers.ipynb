```python
# Creating a dictionary of players and their stats
player_info = {
    "Lionel Messi": {"team": "Inter Miami", "goals": 25},
    "Cristiano Ronaldo": {"team": "Al-Nassr", "goals": 30},
    "Neymar": {"team": "Al-Hilal", "goals": 18},
}

# Adding a new player
player_info["Kylian Mbapp√©"] = {"team": "PSG", "goals": 28}
print("Updated player info:", player_info)


### **Module 3 Solutions**

---

#### Exercise 1: Jump Heights

```python
# Create a Numpy array of jump heights
jump_heights = np.array([35, 42, 38, 45, 40])

# Calculate the average and maximum height
average_height = np.mean(jump_heights)
max_height = np.max(jump_heights)

print("Average height:", average_height)
print("Maximum height:", max_height)

# Add 2 cm to each height
updated_heights = jump_heights + 2
print("Updated heights:", updated_heights)

# Generate an array of 10 angles from 0 to 180 degrees
angles_degrees = np.linspace(0, 180, 10)

# Convert angles to radians
angles_radians = angles_degrees * np.pi / 180

print("Angles in degrees:", angles_degrees)
print("Angles in radians:", angles_radians)



### **Module 4 Solutions**

---

#### Exercise 1: Creating a Dataframe

```python
# Create the dataframe
sports_data = {
    "Sport": ["Basketball", "Tennis", "Soccer"],
    "Players Per Team": [5, 1, 11],
    "Is Team Sport": [True, False, True],
}
df = pd.DataFrame(sports_data)
print("Sports dataframe:\n", df)

# Add the new column
df["Average Duration (minutes)"] = [48, 90, 90]
print("Dataframe with average duration:\n", df)

# Extract specific columns
selected_columns = df[["Sport", "Average Duration (minutes)"]]
print("Extracted columns:\n", selected_columns)


### **Module 5 Solutions**

---

#### Exercise 1: Calculating Statistics with Lists

```python
# Game scores
game_scores = [102, 96, 110, 98, 115]

# Calculate statistics
average_score = sum(game_scores) / len(game_scores)
max_score = max(game_scores)
total_score = sum(game_scores)

print("Average score:", average_score)
print("Maximum score:", max_score)
print("Total score:", total_score)

# Points scored by players
player_points = {"Player A": 20, "Player B": 18, "Player C": 25, "Player D": 15}

# Extract values
points = list(player_points.values())

# Calculate statistics
average_points = sum(points) / len(points)
total_points = sum(points)

print("Average points:", average_points)
print("Total points:", total_points)

import numpy as np

# Sprint times in seconds
sprint_times = np.array([11.2, 10.9, 11.5, 10.8, 11.0])

# Calculate statistics
average_time = np.mean(sprint_times)
fastest_time = np.min(sprint_times)

print("Average sprint time:", average_time)
print("Fastest sprint time:", fastest_time)

import pandas as pd

# Player data
player_data = {
    "Player": ["Player 1", "Player 2", "Player 3", "Player 4"],
    "Goals Scored": [12, 9, 15, 10],
}
df = pd.DataFrame(player_data)

# Calculate statistics
average_goals = df["Goals Scored"].mean()
total_goals = df["Goals Scored"].sum()

print("Average goals scored:", average_goals)
print("Total goals scored:", total_goals)




### **Module 6 Solutions**

---

#### Exercise 1: Line Plot for Player Speed

```python
import matplotlib.pyplot as plt

# Data
time = [0, 1, 2, 3, 4, 5]  # Time in seconds
player_speed = [4.2, 4.5, 5.0, 5.3, 5.1, 4.8]  # Speed in m/s

# Create a line plot
plt.plot(time, player_speed, marker="o")
plt.title("Player Speed Over Time")
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.grid(True)
plt.show()

# Data
teams = ["Team A", "Team B", "Team C"]
points = [45, 50, 40]

# Create a bar chart
plt.bar(teams, points, color="orange")
plt.title("Team Points")
plt.xlabel("Teams")
plt.ylabel("Points")
plt.show()

# Data
time = [0, 1, 2, 3, 4, 5]  # Time in seconds
player_speed = [4.2, 4.5, 5.0, 5.3, 5.1, 4.8]  # Speed in m/s

# Create a customized line plot
plt.plot(time, player_speed, marker="o", linestyle="--", color="red", label="Speed")
plt.title("Customized Player Speed Plot")
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.grid(True)
plt.legend()
plt.show()

# Data
time = [0, 1, 2, 3, 4, 5]  # Time in seconds
player_speed = [4.2, 4.5, 5.0, 5.3, 5.1, 4.8]  # Speed in m/s

# Create a customized line plot
plt.plot(time, player_speed, marker="o", linestyle="--", color="red", label="Speed")
plt.title("Customized Player Speed Plot")
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.grid(True)
plt.legend()
plt.show()


### **Module 7 Solutions**

---

#### Exercise 1: Looping Through a List

```python
# List of player names
player_names = ["Player A", "Player B", "Player C"]

# Loop through the list
for player in player_names:
    print(f"{player} is ready to play!")

# List of game scores
game_scores = [102, 96, 110, 98, 115]

# Loop through the list with index and value
for index, score in enumerate(game_scores):
    print(f"Game {index + 1}: {score} points")

# Dictionary of player points
player_points = {"Player A": 20, "Player B": 18, "Player C": 25}

# Loop through the dictionary
for player, points in player_points.items():
    print(f"{player} scored {points} points.")

# Dictionary of player points
player_points = {"Player A": 20, "Player B": 18, "Player C": 25}

# Loop through the dictionary
for player, points in player_points.items():
    print(f"{player} scored {points} points.")

# Players and their points
players = ["Player A", "Player B", "Player C"]
points = [20, 18, 25]

# Loop through both lists
for player, point in zip(players, points):
    print(f"{player} scored {point} points.")

# Players and their game scores
players = ["Player A", "Player B"]
game_scores = [[20, 25, 30], [15, 18, 22]]  # Scores for 3 games per player

# Nested loop to print scores
for player, scores in zip(players, game_scores):
    print(f"Scores for {player}:")
    for game, score in enumerate(scores, start=1):
        print(f"  Game {game}: {score} points")



### **Module 8 Solutions**

---

#### Exercise 1: Categorizing a Player's Score

```python
# Categorize the player's score
score = 28

if score > 30:
    print("Excellent")
elif 20 <= score <= 30:
    print("Good")
else:
    print("Needs improvement")


# Player scores
scores = [12, 22, 18, 30, 25]

# Check each score
for score in scores:
    if score > 20:
        print(f"{score}: Above threshold")
    else:
        print(f"{score}: Below threshold")


# Players and their ages
player_ages = {"Player A": 22, "Player B": 28, "Player C": 24}

# Check eligibility
for player, age in player_ages.items():
    if age <= 25:
        print(f"{player}: Eligible")
    else:
        print(f"{player}: Not eligible")


import pandas as pd

# Create a dataframe
data = {
    "Player": ["Player A", "Player B", "Player C"],
    "Score": [18, 25, 22],
}
df = pd.DataFrame(data)

# Filter players with scores above 20
filtered_df = df[df["Score"] > 20]
print("Filtered players:\n", filtered_df)




### **Module 9 Solutions**

---

``` python 

# Function with positional, optional, and **kwargs parameters
def calculate_total_score(base_score, bonus=10, **kwargs):
    """
    Calculate the total score of a player, including optional bonus and additional information.

    Parameters:
    base_score (int): The player's base score.
    bonus (int, optional): The bonus score. Defaults to 10.
    **kwargs: Additional keyword arguments (e.g., game_type, season).

    Returns:
    dict: A dictionary containing the total score and additional information.
    """
    total_score = base_score + bonus
    return {"total_score": total_score, **kwargs}

# Example usage
result = calculate_total_score(50, game_type="Final", season="2023")
print(result)

# Function with a docstring to filter scores
def filter_scores(scores, threshold):
    """
    Filter a list of scores to include only values greater than the threshold.

    Parameters:
    scores (list): A list of numerical scores.
    threshold (int): The threshold value.

    Returns:
    list: A list of scores above the threshold.
    """
    return [score for score in scores if score > threshold]

# Example usage
filtered = filter_scores([15, 25, 10, 30, 20], 20)
print("Filtered scores:", filtered)


# Function using None as a default parameter
def player_eligibility(age, experience=None):
    """
    Determine a player's eligibility based on age and experience.

    Parameters:
    age (int): The player's age.
    experience (int, optional): The player's years of experience. Defaults to None.

    Returns:
    bool: True if eligible, False otherwise.
    """
    if experience is None:
        return age < 25
    return age < 25 and experience >= 2

# Example usage
print(player_eligibility(24))  # Uses default experience=None
print(player_eligibility(24, experience=3))
print(player_eligibility(28, experience=3))


```
